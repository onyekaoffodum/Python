```python
# Import libraries 

from bs4 import BeautifulSoup
import requests
import time
import datetime
import smtplib
```

```python
# Connect to Website

URL = 'https://www.jumia.com.ng/catalog/?q=handbags+for+ladies'

headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36 Edg/121.0.0.0", "Accept-Encoding":"gzip, deflate, br", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}
```

```python
# Pulling in data

page = requests.get (URL, headers= headers)

Soup1 = BeautifulSoup (page.content, 'html.parser')

Soup2 = BeautifulSoup (Soup1.prettify(), 'html.parser')

name = Soup2.find(class_= 'name')

price = Soup2.find(class_= 'prc')

print(price)

print(name)
```

```python
# Iterate for the list of prices

price_elements = Soup2.find_all(class_='prc')

for price_element in price_elements:
    price = price_element.get_text().strip()[2:] # removing the first two characters
    print(price)
```

```python
# Iterate for the list of names

name_elements = Soup2.find_all(class_='name')

for name_element in name_elements:
    name = name_element.get_text().strip()
    print(name)
```

```python
# Create csv and type headers and data into the file

import csv

header = ['Name', 'Price']

# List of name_elements and price_elements
name_elements = Soup2.find_all(class_='name')
price_elements = Soup2.find_all(class_='prc')

# Now we are importing data into the csv
with open('JumiaWebScraperDatasetPart2.csv', 'w', newline='', encoding='UTF8') as f:
    writer = csv.writer(f)
    writer.writerow(header)

    # Zip the name_elements and price_elements together for iteration
    for name_element, price_element in zip(name_elements, price_elements):
        name = name_element.get_text().strip()
        price = price_element.get_text().strip()[2:]  
        writer.writerow([name, price])
```
